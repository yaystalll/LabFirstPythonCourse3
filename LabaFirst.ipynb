```python
import csv
import numpy as np

def read_column_from_csv(filename, column_name):
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)
        column_index = headers.index(column_name)
        column_data = np.array([int(row[column_index]) for row in reader])
    return column_data

aircraft_losses = read_column_from_csv('russia_losses_equipment.csv', 'aircraft')

daily_aircraft_losses = np.diff(aircraft_losses * -1)

top_3_losses = np.sort(daily_aircraft_losses)[-3:]

print("Три найбільші добові втрати літаків:", top_3_losses)
```

```python
import csv
import numpy as np

def read_data_with_dates(filename, column_name):
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)
        date_index = headers.index("date")
        column_index = headers.index(column_name)
        
        data = []
        dates = []
        for row in reader:
            dates.append(row[date_index])
            data.append(int(row[column_index]))
        
        return np.array(dates), np.array(data)
    
def filter_summer_2022(dates, data):
    summer_2022_start = '2022-06-01'
    summer_2022_end = '2022-08-31'
    
    summer_mask = (dates >= summer_2022_start) & (dates <= summer_2022_end)
    
    return data[summer_mask]

dates, aircraft_losses = read_data_with_dates('russia_losses_equipment.csv', 'aircraft')

summer_2022_losses = filter_summer_2022(dates, aircraft_losses)

total_summer_2022_losses = summer_2022_losses[0] - summer_2022_losses[-1]

print(f"Загальна кількість збитих літаків влітку 2022 року: {total_summer_2022_losses}")
```

```python
import csv
import numpy as np

def read_aircraft_losses(filename, column_name):
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)
        column_index = headers.index(column_name)
        
        data = []
        for row in reader:
            data.append(int(row[column_index]))
        
        return np.array(data)

def calculate_average_losses(data, num_days):
    last_losses = data[:num_days]
    daily_losses = np.diff(last_losses * -1)
    
    return np.mean(daily_losses)

aircraft_losses = read_aircraft_losses('russia_losses_equipment.csv', 'aircraft')
average_losses_last_300_days = calculate_average_losses(aircraft_losses, 300)

print(f"Середнє значення втрат авіації за останні 300 днів: {average_losses_last_300_days:.2f}")
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

def read_data(filename, date_col, aircraft_col):
    dates = []
    aircraft_losses = []
    
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)
        date_index = headers.index(date_col)
        aircraft_index = headers.index(aircraft_col)
        
        for row in reader:
            date_str = row[date_index]
            date_obj = datetime.strptime(date_str, "%Y-%m-%d")
            dates.append(date_obj)
            aircraft_losses.append(int(row[aircraft_index]))
    
    return np.array(dates), np.array(aircraft_losses)

def plot_aircraft_losses(dates, losses):
    start_date = datetime(2022, 2, 25)
    end_date = datetime(2023, 2, 25)
    
    mask = (dates >= start_date) & (dates <= end_date)
    filtered_dates = dates[mask]
    filtered_losses = losses[mask]

    daily_losses = np.diff(filtered_losses * -1)

    plt.figure(figsize=(16, 32), dpi=200)
    
    plt.plot(filtered_dates[1:], daily_losses, linestyle='--', color='b', label='Добові втрати літаків')

    plt.xlabel('Дата', fontsize=14)
    plt.ylabel('Втрати літаків', fontsize=14)

    plt.yticks(np.arange(0, 26, 1))
    
    plt.title('Втрати літаків у перший рік війни (Від 2022-02-25 до 2023-02-25)', fontsize=16)

    plt.legend(loc='upper right')

    plt.grid(True, which='both', linestyle=':', linewidth=0.5, color='gray')

    plt.savefig('aircraft_losses_first_year.png', format='png')
    
    plt.show()

filename = 'russia_losses_equipment.csv'
dates, aircraft_losses = read_data(filename, 'date', 'aircraft')

plot_aircraft_losses(dates, aircraft_losses)
```
